```python
import numpy as np
import pandas as pd
```

```python
## using dataset https://www.kaggle.com/datasets/jutrera/stanford-car-dataset-by-classes-folder
```

```python
train_dir=r'C:\Users\abalu\Desktop\car_data\car_data\train'
test_dir=r'C:\Users\abalu\Desktop\car_data\car_data\test'
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
import os
```

```python
train_gen=ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
)
test_gen=ImageDataGenerator(
rescale=1./255)
```

```python
#using resnet : https://keras.io/api/applications/resnet/#resnet50-function
from tensorflow.keras.applications.resnet50 import ResNet50
```

```python
resnet=ResNet50(input_shape=[150,150] +[3],weights='imagenet',include_top=False)
```

```python
for layer in resnet.layers:
    layer.trainable=False
```

```python
from glob import glob
```

```python
folder=glob(train_dir +'/*')
```

```python
len(folder)
```

```python
import tensorflow.keras as k
```

```python
model=k.layers.Flatten()(resnet.output)
```

```python
final=k.layers.Dense(len(folder),activation='softmax')(model)
```

```python
model=k.Model(inputs=resnet.input,outputs=final)
```

```python
model.summary()
```

```python
model.compile(
loss='categorical_crossentropy',
optimizer='adam',
metrics=['accuracy']
)
```

```python
train=train_gen.flow_from_directory(
train_dir,target_size=(150,150),
    batch_size=20,
    class_mode='categorical'
)
```

```python
test=test_gen.flow_from_directory(
train_dir,target_size=(150,150),
    batch_size=20,
    class_mode='categorical'
)
```

```python
modelx=model.fit_generator(train,validation_data=test,epochs=60,steps_per_epoch=len(train),validation_steps=len(test))
```
