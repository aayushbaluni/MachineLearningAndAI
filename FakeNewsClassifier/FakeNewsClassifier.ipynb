```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("fake_or_real_news.csv")
```

```python
df
```

```python
df=df.drop("Unnamed: 0",axis=1)
```

```python
df
```

```python
df['comb']=df['title']+" "+df['text']
```

```python
df
```

```python
y=pd.get_dummies(df['label']).values
```

```python
y=y[:,1]
```

```python
y
```

```python
data=df['title']
```

```python
data
```

```python
data.replace('[^a-zA-Z]',' ',regex=True,inplace=True)
```

```python
data
```

```python
news=[]
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
ps=PorterStemmer()
```

```python
for i in range(0,len(data)):
    r=data[i]
    r.lower()
    r.split()
    r=[ps.stem(word)for word in r if word not in stopwords.words('english')]
    r=''.join(r)
    news.append(r)
```

```python
news
```

```python
from tensorflow.keras.preprocessing.text import one_hot
vocab_size=100000
```

```python
hot_text=[one_hot(word,vocab_size) for word in news]
```

```python
hot_text
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
emb_len=0
for i in range( 0, len(hot_text)):
    emb_len=max(emb_len,len(hot_text[i]))
emb_len
```

```python
pad=pad_sequences(hot_text,padding='pre',maxlen=emb_len)
```

```python
pad
```

```python
emb_features=100
import tensorflow
```

```python
model=tensorflow.keras.models.Sequential()
```

```python
model.add(tensorflow.keras.layers.Embedding(vocab_size,emb_features,input_length=emb_len))
```

```python
model.add(tensorflow.keras.layers.Dropout(0.5))
```

```python
model.add(tensorflow.keras.layers.LSTM(500))
```

```python
model.add(tensorflow.keras.layers.Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',metrics=['acc'])
```

```python
pad
```

```python
xtrain=np.array(pad)
```

```python
ytrain=np.array(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(xtrain,ytrain,test_size=0.2)
```

```python
x_test.shape
```

```python
x_train.shape
```

```python
model1=model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=20,batch_size=64)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```
